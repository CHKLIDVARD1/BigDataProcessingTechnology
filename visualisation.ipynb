```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
data = np.load('average_ratings.npy')
df = pd.DataFrame(data)
df.index = ['waffle iron french toast', 'zwetschgenkuchen bavarian plum cake', 'lime tea']
df = df.T
```

```python
fig, ax = plt.subplots(figsize = (10,7))
y1 = df['waffle iron french toast'].values
y2 = df['zwetschgenkuchen bavarian plum cake'].values
y3= df['lime tea'].values
x = df.index.values
fig.suptitle('    ', size = 20)
ax.plot(x, y1, color = '#b00000', label = 'waffle iron french toast')
ax.plot(x, y2, color = '#19ff19', label = 'zwetschgenkuchen bavarian plum cake')
ax.plot(x, y3, color = '#ebdb00', label = 'lime tea')
ax.set_xlabel(' ', size = 12)
ax.set_ylabel(' ', size = 12)
ax.legend();

```

```python
reviews = pd.read_csv('reviews_sample.csv')
recipes = pd.read_csv('recipes_sample.csv')
```

```python
recipes_short = recipes[recipes.minutes < 5]
recipes_average = recipes[(recipes.minutes > 4) & (recipes.minutes < 50)]
recipes_long = recipes[recipes.minutes > 50]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10,7), dpi = 80)
short_size = recipes_short.shape[0]
average_size = recipes_average.shape[0]
long_size = recipes_long.shape[0]
avg_steps_short = recipes_short['n_steps'].mean()
avg_steps_average = recipes_average['n_steps'].mean()
avg_steps_long = recipes_long['n_steps'].mean()
df = pd.DataFrame([avg_steps_short, avg_steps_average, avg_steps_long])
df.index = ['', '', '']
df = df.rename(columns = {0: '  '})
df[' '] = [short_size, average_size, long_size]
df['  '].plot(kind = 'bar', ax=axes[0])
axes[0].set_xlabel(' ', size = 15)
axes[0].set_ylabel(' ', size = 15)

df[' '].plot(kind = 'pie', ax=axes[1]);
axes[1].set_title("  ", size = 15);
```

```python
reviews = reviews[(reviews.date >= '2008-01-01') & (reviews.date < '2010-01-01')]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10,7), dpi = 80)
reviews2008 = reviews[reviews.date < '2009-01-01']
reviews2009 = reviews[reviews.date > '2008-12-31']
reviews2008['rating'].plot.hist(ax=axes[0])
reviews2009['rating'].plot.hist(ax=axes[1])
fig.subplots_adjust(wspace = 0.3)
fig.suptitle("    2008  2009 ", size = 20);
```

```python
recipes['Duration'] = recipes.minutes
recipes.loc[(recipes.minutes < 5), 'Duration'] = ''
recipes.loc[((recipes.minutes > 4) & (recipes.minutes < 50)), 'Duration'] = ''
recipes.loc[(recipes.minutes >= 50), 'Duration'] = ''
sns.displot(recipes, x = 'n_steps', y = 'n_ingredients', hue = 'Duration').set(title = '  n_steps  n_ingredients');
```

```python
big_data = pd.merge(recipes, reviews, left_on  = 'id', right_on = 'recipe_id', how = 'inner')
```

```python
cols = np.array(["minutes", "n_steps", "n_ingredients", "rating"])
df4 = big_data.loc[:, big_data.columns.isin(cols)]
sns.heatmap(df4.corr(), annot = True, cmap = 'YlOrRd').set(title = '     recipes  reviews');
```
